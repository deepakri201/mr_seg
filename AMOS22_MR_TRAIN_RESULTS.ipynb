```python
# Save output in a variable `df`
# %%bigquery mrsegmentator_results_df --project idc-external-018
# SELECT *
#   FROM `idc-external-018.synthseg_exp4B.testing_amos22_mr_train_mrsegmentator`
!wget -N https://raw.githubusercontent.com/deepakri201/mr_seg/main/amos22_mr_train_split_mrsegmentator_results.csv
amos22_mr_train_split_mrsegmentator_results = pd.read_csv("amos22_mr_train_split_mrsegmentator_results.csv")
```

```python
mrsegmentator_results_df["method"] = 'MRSegmentator'
mrsegmentator_results_df = mrsegmentator_results_df[mrsegmentator_results_df.segment.isin(["liver","spleen","kidney_left", "kidney_right"])]
mrsegmentator_results_df = mrsegmentator_results_df[["caseID", "segment", "dsc", "hsdff_95", "method"]]
```

```python
mrsegmentator_results_df.head()
```

```python
df_concat_abdo = pd.concat([synthseg_results_df, totalsegmri_results_df, mrsegmentator_results_df])
df_concat_abdo = df_concat_abdo[df_concat_abdo.segment.isin(["spleen", "liver", "kidney_left", "kidney_right"])]
```

```python
fig = px.box(df_concat_abdo, x="segment", y="dsc", color="method", width=900, height=500,
             hover_data=["caseID"], points="all")
fig.update_traces(quartilemethod="exclusive") # or "inclusive", or "linear" by default
# fig.update_traces(boxpoints=False)
fig.show()
```
